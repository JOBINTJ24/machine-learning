```python
import pandas as pd
import numpy as np
```

```python
print(pd.Series(np.array(['a','b','c','d']), index=['first','second','third','fourth']))
```

```python
print(pd.Series(50, index=[0,1,2,3,4,5]))
```

```python
print(pd.Series({1:10, 2:20, 3:30, 4:40, 5:50, 6:60}))
```

```python
demo=pd.Series([1,2,3,4,5,6,7,8])
print('multiply',demo*2)
print('square', demo**2)
```

```python
print(demo.head(3))
print(demo.tail(3))
```

```python
print(demo.loc[:4])
print(demo.iloc[:3])
print(demo[2:5])
```
